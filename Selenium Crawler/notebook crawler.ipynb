```python
import pandas as pd
dfMaster = pd.DataFrame(columns = ["Team", "Home","Draw","Away"])
className = """gl-MarketGroupContainer"""

els = driver.find_elements_by_class_name(className)

count = 0

for el in els:
    df = pd.DataFrame(columns = ["Team", "Home","Draw","Away"])

    elText = el.text
    if elText == "":
        continue
    className = """rcl-ParticipantFixtureDetails_TeamAndScoresContainer"""

    TeamELs = el.find_elements_by_class_name(className)

    
    className = """sgl-ParticipantOddsOnly80_Odds"""

    OddELs = el.find_elements_by_class_name(className)

    odds = []
    onematchOdd = []
    count = 0
    for el in OddELs:
        onematchOdd.append(el.text)
        count+=1
        if count == len(TeamELs):
            odds.append(onematchOdd)
            count = 0
            onematchOdd = []
    
    TeamList = [el.text.split("\n") for el in TeamELs if not el.text.isdigit()]
    
    df["Team"] = TeamList
    df["Home"] = odds[0]
    df["Draw"] = odds[1]
    df["Away"] = odds[2]
    
    dfMaster = dfMaster.append(df)
    
dfMaster
```

```python
def sliptTeam(row):
    stringTeam = row["Team"] 
    teamList = list(stringTeam)
#     print(teamList)
    row["HomeTeam"] = teamList[0]
    row["AwayTeam"] = teamList[1]
    return row

dfMaster = dfMaster.apply(sliptTeam,axis =1)
dfMaster.drop(["Team"],axis =1,inplace = True)
dfMaster = dfMaster[["HomeTeam","AwayTeam","Home","Draw","Away"]]
dfMaster.dropna()
```

```python
import numpy as np
dfMaster = dfMaster.replace(r'^\s*$', np.nan, regex=True)
dfMaster = dfMaster.dropna()
dfMaster
```

```python
def addDecimalOdd(row):
    listOdd = []
    listOdd.append(row["Home"])
    listOdd.append(row["Draw"])
    listOdd.append(row["Away"])
    
    returnList = []
    for el in listOdd:
        if "+" in el:
            returnEl = int(el.replace("+", ""))
            returnEl = returnEl/100 + 1
        else:
            returnEl = int(el.replace("-", ""))
            returnEl = 1 - (100/ -returnEl)
        returnList.append(returnEl)
    row["HomeDecimal"] = returnList[0]
    row["DrawDecimal"] = returnList[1]
    row["AwayDecimal"] = returnList[2]
    
    return row
dfMaster = dfMaster.apply(addDecimalOdd,axis =1)
dfMaster
```
